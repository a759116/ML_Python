```python
#Hierarchical Clustering
#Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

#Loading the dataset using pandas
dataset = pd.read_csv('shopping.csv')
X = dataset.iloc[:,[3,4]].values


#Finding the optimal number of clusters using the dendrogram
import scipy.cluster.hierarchy as sch
dendro= sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distances')
plt.grid(axis='both', alpha=0.75)
plt.show()

#Hierarchically clustering our dataset
from sklearn.cluster import AgglomerativeClustering
aggloClusObj= AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_result= aggloClusObj.fit_predict(X)

# Visualizing the clusters
plt.scatter(X[y_result== 0, 0], X[y_result== 0, 1], s = 100, c = 'red', label = 'High Income Low Spending')
plt.scatter(X[y_result== 1, 0], X[y_result== 1, 1], s = 100, c = 'blue', label = 'Medium Income Medium Spending')
plt.scatter(X[y_result== 2, 0], X[y_result== 2, 1], s = 100, c = 'purple', label = 'High Income High Spending')
plt.scatter(X[y_result== 3, 0], X[y_result== 3, 1], s = 100, c = 'cyan', label = 'Low Income High Spending')
plt.scatter(X[y_result== 4, 0], X[y_result== 4, 1], s = 100, c = 'magenta', label = 'Low Income Low Spending')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.legend(loc=9, bbox_to_anchor=(0.5, -0.2))
plt.grid(axis='both', alpha=0.75)
plt.show()
```
