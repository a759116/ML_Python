Import the dataset and run the Apriori Rule Learning algorithm. 

```python
# Assignment 4
# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

dataset = pd.read_csv('titanic.csv')
dataset.head()
```

```python
records = []  
for i in range(0, len(dataset)):
    records.append([str(dataset.values[i,j]) for j in range(0, len(dataset.columns))])
    

# Training Apriorion the dataset
from apyori import apriori
association_rules= apriori(records, min_support=0.01, min_confidence=0.3, min_lift=1, max_length=4)  
association_rules= list(association_rules) 
```

```python
sum = 0
for r in range(0, len(association_rules)):
    sum += len(association_rules[r][2])
print(sum)
```

Display the rules you get in a proper format where you first show the rule:
{A}{B} followed by Support, Confidence, Lift below that.

```python
rule_seq = 0
for item in association_rules:
    for j in range(0, len(item[2])):
        a = item[2][j][0]
        b = item[2][j][1]
        rule_seq = rule_seq + 1
        print(rule_seq)
        print("Rule: " + str(list(a)) + " -> " + str(list(b)))
        #second index of the inner list
        print("Support: " + str(item[1]))
        #third index of the list located at 0th
        #of the third index of the inner list
        print("Confidence: " + str(item[2][j][2]))
        print("Lift: " + str(item[2][j][3]))
        print("=====================================")
```

Pruning the rules based on consequents: Write a function that takes as parameters: support, confidence, and a consequent attribute.
o e.g. If the consequent attribute parameter passed is “Sex”, then your function should only display rules where the consequent has “Sex” only.
▪ If Survived=Yes and Class=3rd  Sex=Male (should be displayed)
▪ If Survived=Yes and Class=3rd  Sex=Male and Age=Adult (should not be displayed)
o e.g. If the consequent attribute parameter passed is “Sex” and “Age”, then your function should only display rules where the consequent has “Sex” and “Age”.
▪ If Survived=Yes and Class=3rd  Sex=Male (should not be displayed)
▪ If Survived=Yes and Class=3rd  Sex=Male and Age=Adult (should be displayed)

```python
def prune_rules_on_consequents(support, confidence, consequents):
    association_rules= apriori(records, min_support=support, min_confidence=confidence, min_lift=1, max_length=4)  
    association_rules= list(association_rules)

    for item in association_rules:
        for j in range(0, len(item[2])):
            a = item[2][j][0]
            b = item[2][j][1]
            if(consequents == list(b)):
                print("Rule: " + str(list(a)) + " -> " + str(list(b)))
                #second index of the inner list
                print("Support: " + str(item[1]))
                #third index of the list located at 0th
                #of the third index of the inner list
                print("Confidence: " + str(item[2][j][2]))
                print("Lift: " + str(item[2][j][3]))
                print("=====================================")
```

```python
prune_rules_on_consequents(0.01,0.3,['Female'])
```

Pruning redundant rules: Write a function that takes support and confidence as parameters and prunes redundant rules while displaying the rules.
o e.g. Rule 1: If Class=2nd and Age=Child  Survived = Yes
Rule 2: If Class=2nd and Age=Child and Sex=Female  Survived = Yes (redundant)
Rule 2 does not give us any additional information and hence is a redundant rule. Your function should only display non-redundant rules.

```python
def prune_redundant_rules(support, confidence):
    association_rules= apriori(records, min_support=support, min_confidence=confidence, min_lift=1, max_length=4)  
    association_rules= list(association_rules)
    
    rules_result = {}
    for item in association_rules:
        for j in range(0, len(item[2])):
            a = item[2][j][0]
            b = item[2][j][1]
            if str(list(b)) in rules_result:
                #if a is aubset of any of the antecedents of b
                #replace that antecedent with a, otherwise add a 
                current_antecedents = rules_result.get(str(list(b)))
                new_antecedents = list(current_antecedents)
                
                """
                x = list(b)
                if (x[0] == 'Female'):
                    print("PRE  ", str(list(b)), "\t", new_antecedents, "\t", str(list(a)))
                """
                
                setNew = set(list(a))
                isNewSuperSet = False
                
                for current_antecedent in current_antecedents:
                    setExisting = set(current_antecedent)
                    
                    if (setNew.issuperset(setExisting)):
                        isNewSuperSet = True
                    elif (setNew.issubset(setExisting)):
                        new_antecedents.remove(list(current_antecedent))
                    else:
                        pass

                """
                if (x[0] == 'Female'): 
                    print("POST  ", str(list(b)), "\t", rules_result[str(list(b))], "\t", new_antecedents)
                """
                
                if(not isNewSuperSet):
                    new_antecedents.append(list(a))
                rules_result[str(list(b))] = new_antecedents
 
            else:
                rules_result[str(list(b))] = [list(a)]
            
    for consequent, antecedents in rules_result.items():
        for antecedent in antecedents:
            print("Rule: " + str(antecedent) + " -> " + str(consequent))
```

```python
prune_redundant_rules(0.01,0.5)
```
